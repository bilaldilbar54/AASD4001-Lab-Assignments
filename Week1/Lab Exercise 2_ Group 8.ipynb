```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# # of hours a AI student spends on their courses in a month
np.random.seed(10)
no_hours = np.random.normal(40.0, 10.0, 10000)

plt.hist(no_hours, 50)
plt.show()

```

```python
# Calculate and report the mean, std, and var of the data shown above.

mean_hours = np.mean(no_hours)
std_hours = np.std(no_hours)
var_hours = np.var(no_hours)
print("Average number of hours AI student studying is:",mean_hours)
print("Standard deviation of student's studying hours is :",std_hours)
print("Variance of student's studying hours is:",var_hours)

```

```python
# What is the theoretical relationship between std and var? Do the numbers reperted in the previous step agree with your expectation?
```


Answer: The relationship between standard deviation and variance is that the standard deviation is square root of variance.
