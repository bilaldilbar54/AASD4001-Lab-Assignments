The process below shows how to implement a simple linear regression analysis for a vehicle pricing scenario. Re-run it and take note of the R-sqaured value. Then. change the std value in np.random.normal to 1, 5 and 10. Observe the change in R-sqaured. How do you explain the change in R-sqaured value?

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

vehicle_age = np.random.uniform(0,10,1000)
vehicle_price = 70000 - (vehicle_age + np.random.normal(0, 1, 1000)) * 3000

vehicle_age = vehicle_age.reshape(-1, 1)  #reshape for single feature, use (1,-1) for single sample
vehicle_price = vehicle_price.reshape(-1, 1)

plt.scatter(vehicle_age, vehicle_price)
```

```python
reg = LinearRegression().fit(vehicle_age, vehicle_price)
reg.score(vehicle_age, vehicle_price)
```

Observation when std value set to 1 :

When std = 1, the relationship between vehicle age and pricing will be tighter, basically it is close to linear, and we can see that the graph is just like a thick line. We found that vehicle age is inversely proportional to vehicle price. We can find this relationship between these two parameter easily. For std =1, the R squared value is high (near 0.9), which is good for the regression model. An R squared of 0.9 reveals that 90% of the variability observed in the target variable is explained by the regression model.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

vehicle_age = np.random.uniform(0,10,1000)
vehicle_price = 70000 - (vehicle_age + np.random.normal(0, 10, 1000)) * 3000

vehicle_age = vehicle_age.reshape(-1, 1)  #reshape for single feature, use (1,-1) for single sample
vehicle_price = vehicle_price.reshape(-1, 1)

plt.scatter(vehicle_age, vehicle_price)
```

```python
reg = LinearRegression().fit(vehicle_age, vehicle_price)
reg.score(vehicle_age, vehicle_price)
```

Observation when std value set to 10 :

When std = 10, the relationship between vehicle age and pricing will be difficult to observe, we are not able to observe the graph including the tight zone, as the point is scattered. The inverse proportion is not clear, even not enough to deduce the relationship easily. For std =10, the R squared value is very low (near 0.09), which is very bad for the regression model compared to std = 1 or 5. An R squared of 0.09 reveals that 9% of the variability observed in the target variable is explained by the regression model. This R-squared figure is a bad sign for predictive models, it is hard to predict the relationship between vehicle price and vehicle age.
